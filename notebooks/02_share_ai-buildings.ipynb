```python
import os

import geopandas as gpd
import pandas as pd
```

```python
# OSM AI-generated data stats
ai_df = gpd.read_file("data/blds_world-states-2024.geojson")

columns_to_keep = [
    "ISO_A3",
    "ai_blds_2024",
    "blds_total",
    "ai_percentage",
    "ai_per_roun"
]
ai_df.drop(
    ai_df.columns.difference(columns_to_keep),
    axis=1, 
    inplace=True
)
ai_df.set_index("ISO_A3", inplace=True)

ai_df["non_ai_blds_2024"] = ai_df["blds_total"] - ai_df["ai_blds_2024"]

ai_df.reset_index(inplace=True)

display(ai_df)
```

```python
# sort values by number of ai blds
ai_df.sort_values(by='ai_blds_2024', ascending=False, inplace=True)

# compute sum of all ai blds
ai_df['ai_blds_2024_total'] = ai_df['ai_blds_2024'].sum()

# select first 20 countries and make a copy to avoid SettingWithCopyWarning
ai_df_20_sel = ai_df.iloc[0:21].copy()

# compute the share of ai blds
ai_df_20_sel.loc[:, 'ai_blds_share'] = (ai_df_20_sel['ai_blds_2024'] / ai_df_20_sel['ai_blds_2024_total']) * 100

display(ai_df_20_sel)
```
